# 서울 20,21,22년도

```python
# 서울 20,21,22

# 현재 작업 디렉터리에 있는 chromedriver 파일을 가리킵니다.
s = Service('./chromedriver.exe')


driver = webdriver.Chrome(service=s)

# 지정한 웹 페이지 열기
driver.get("https://tmacs.kotsa.or.kr/web/TG/TG300/TG3050/Tg3050.jsp?mid=S1830#")

time.sleep(5)

# 광역 자치단체 선택 박스를 찾아 '서울'을 선택합니다.
select_seoul = driver.find_element(By.XPATH, '//*[@id="sido"]/option[2]')
select_seoul.click()

# '조회' 버튼 클릭하기
search_button = driver.find_element(By.XPATH, '//*[@id="content"]/div[2]/div[1]/div[2]/a')
search_button.click()

# 로딩 시간 고려하여 10초 대기하기
time.sleep(10)

#스크롤 내리기
driver.execute_script("window.scrollTo(0, 380)")
#데이터 로딩 기다리기
time.sleep(5)
html = driver.page_source
soup= BeautifulSoup(html, "html.parser")
#리스트에 이름을 넣기
spot_list = soup.find_all("span", "rMateH5__DataGridItemRenderer rMateH5__DataGridColumn25")
#rMateH5__Content71 > span:nth-child(4)
time.sleep(5)

a=0
#총 데이터수
count=0

while(True):
    for i in range(len(spot_list)):
        j = (a+1)*2
        sopt = spot_list[a].get_text()
        #상세보기
        driver.find_element(By.XPATH,'//*[@id="rMateH5__Content77"]/div[%d]/span/img' %j).click()
        #time.sleep(3) #데이터 로딩 기다리기
        #창 전환
        driver.switch_to.window(driver.window_handles[-1])
        driver.get_window_position(driver.window_handles[-1])

        #세부조회
        driver.find_element(By.XPATH,'//*[@id="new_popup"]/div[2]/div[1]/ul/li[3]/a').click()
        time.sleep(15) #데이터 로딩 기다리기
        #엑셀 다운로드
        driver.find_element(By.XPATH,'//*[@id="new_popup"]/div[2]/div[3]/a[2]').click()
        time.sleep(4)

        driver.close()

        # 파일 이름 바꾸기 및 이동
        download_path = 'C:\\Users\\songhow53\\Downloads'
        new_folder_path = 'D:\\리스크관리공모전\\서울202122_2'

        filename = max([download_path + '\\' + f for f in os.listdir(download_path)], key=os.path.getctime)

        new_filename = sopt + '.xls'

        shutil.move(filename, os.path.join(new_folder_path, new_filename))

        #원래 창으로
        driver.switch_to.window(driver.window_handles[0])
        driver.get_window_position(driver.window_handles[0])

        if(a==12):
            #스크롤 내리기

            scroll = driver.find_element(By.XPATH,'//*[@id="grid1"]/div/div/div[2]')

            # ActionChains생성
            action = ActionChains(driver)

            # 클릭하고 잡기
            action.click_and_hold(scroll).perform()

            # 마우스 내리기
            for i in range(21):
                action.send_keys(Keys.ARROW_DOWN)

            # 마우스 놓아주기
            action.release(scroll).perform()
            html = driver.page_source
            soup= BeautifulSoup(html, "html.parser")
            spot_list = soup.find_all("span", "rMateH5__DataGridItemRenderer rMateH5__DataGridColumn25")
            a=-1

        a += 1
        count += 1
    
    if(count>699):
        break
        
# 브라우저 (드라이버) 종료
driver.close()
```

```python
import pandas as pd
import os

# 병합할 액셀 파일들이 있는 디렉토리 경로 설정
directory1 = 'D:\\리스크관리공모전\\서울171819'
directory2 = 'D:\\리스크관리공모전\\서울202122'

# 모든 액셀 파일들의 경로 가져오기 (.xls 확장자로 가정)
file_paths1 = [os.path.join(directory1, file) for file in os.listdir(directory1) if file.endswith('.xls')]
file_paths2 = [os.path.join(directory2, file) for file in os.listdir(directory2) if file.endswith('.xls')]

# 데이터를 저장할 빈 데이터프레임 생성
merged_df = pd.DataFrame()

# 첫 번째 디렉토리의 액셀 파일들 병합
for path in file_paths1:
    df = pd.read_excel(path)
    df['사고구역'] = os.path.basename(path).split('.')[0]  # 사고구역 컬럼에 파일명 추가 (확장자 제외)
    merged_df = pd.concat([merged_df, df.iloc[2:]], axis=0)

# 두 번째 디렉토리의 액셀 파일들 병합
for path in file_paths2:
    df = pd.read_excel(path)
    df['사고구역'] = os.path.basename(path).split('.')[0]  # 사고구역 컬럼에 파일명 추가 (확장자 제외)
    merged_df = pd.concat([merged_df, df.iloc[2:]], axis=0)

# 결과 확인
print(merged_df.head())
```

```python
import pandas as pd
import os

# 병합할 액셀 파일들이 있는 디렉토리 경로 설정
directory1 = 'D:\\리스크관리공모전\\경기171819'

# 모든 액셀 파일들의 경로 가져오기 (.xls 확장자로 가정)
file_paths1 = [os.path.join(directory1, file) for file in os.listdir(directory1) if file.endswith('.xls')]

# 데이터를 저장할 빈 데이터프레임 생성
merged_df = pd.DataFrame()

# 첫 번째 디렉토리의 액셀 파일들 병합
for path in file_paths1:
    df = pd.read_excel(path)
    df['사고구역'] = os.path.basename(path).split('.')[0]  # 사고구역 컬럼에 파일명 추가 (확장자 제외)
    merged_df = pd.concat([merged_df, df.iloc[2:]], axis=0)
    
merged_df
```

```python
merged_df.to_excel('경기171819.xlsx', index=False)
```

```python
import pandas as pd

# 파일 경로 설정
file_path1 = r"D:\리스크관리공모전\경기이륜만.xlsx"
file_path2 = r"D:\리스크관리공모전\경기모든사고지점위경도.xlsx"

# 엑셀 파일에서 데이터 읽어오기
df1 = pd.read_excel(file_path1)
df2 = pd.read_excel(file_path2)

# df2에서 '사고구역', 'Latitude', 'Longitude' 컬럼만 선택하여 새로운 데이터프레임 생성
df2_selected = df2[['사고구역', 'Latitude', 'Longitude']]

# df1과 df2_selected를 사고구역 기준으로 병합하기 (left join)
df_merged = pd.merge(df1, df2_selected, on='사고구역', how='left')

df_merged
```

```python
df_merged.to_excel('경기이륜만.xlsx', index=False)
```
